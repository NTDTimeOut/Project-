```python
import numpy as np
import pandas as pd
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
# Đường dẫn tới thư mục chứa các file .csv
data_train_dir = '../Data_set/Data_train/'
data_test_dir = '../Data_set/Data_test/'

# features = ["Gx", "Gy", "Gz", "Ax", "Ay", "Az"]

# Load dữ liệu từ các file .csv
def load_data(path):
    data = []
    labels = []
    for file_name in os.listdir(path):
        if file_name.endswith("_0.csv"):
            labels.append(0)  # Nhãn nữ
        elif file_name.endswith("_1.csv"):
            labels.append(1)  # Nhãn nam
        else:
            continue
        
        file_path = os.path.join(path, file_name)
        df = pd.read_csv(file_path)
        # flattened_data = df[features].values.flatten()
        data.append(df)  # Thêm dữ liệu từ file .csv vào danh sách data
    return np.array(data), np.array(labels)

# Lấy nhãn từ tên file .csv
# def get_label(file_name):
#     # Implement logic để lấy nhãn từ tên file
#     # Ví dụ: nếu tên file chứa từ "male" thì nhãn là 1, ngược lại là 0
#     if '_1' in file_name:
#         return 1
#     else:
#         return 0

# Xây dựng mô hình Deep Learning
def build_model(input_shape):
    model = Sequential()
    model.add(Dense(64, activation='relu', input_shape=input_shape))
    model.add(Dropout(0.1))
    model.add(Dense(64, activation='tanh'))
    model.add(Dense(64, activation='swish'))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    return model

# Load dữ liệu
data_train, labels_train = load_data(data_train_dir)
data_test, labels_test = load_data(data_test_dir)

# Mở rộng chiều của labels_train và labels_test
labels_train = np.expand_dims(labels_train, axis=1)
labels_test = np.expand_dims(labels_test, axis=1)


# Xây dựng mô hình
model = build_model(input_shape=data_train[0].shape)

# Biên dịch mô hình
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'],  run_eagerly=True)

tf.config.run_functions_eagerly(True)

# Huấn luyện mô hình
model.fit(data_train, labels_train, epochs=25, batch_size=32, validation_data=(data_test, labels_test))

# Đánh giá mô hình trên tập kiểm tra
loss, accuracy = model.evaluate(data_test, labels_test)
print('Loss:', loss)
print('Accuracy:', accuracy)
```
